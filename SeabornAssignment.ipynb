Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter Plot
Use: Visualizing the relationship between two numerical variables. Useful for identifying patterns, clusters, or correlations between the variables.

Histogram
Use: Displaying the distribution of a single numerical variable. Helpful for understanding the data's spread, central tendency, and presence of outliers.

Bar Plot
Use: Comparing categorical data by showing the distribution of a numerical variable across different categories. Ideal for visualizing counts or aggregations.

Box Plot (or Box-and-Whisker Plot)
Use: Summarizing the distribution of a numerical variable across different levels of a categorical variable. Useful for identifying outliers, range, and quartile information.

Heatmap
Use: Visualizing the correlation matrix of numerical variables in a dataset. Great for identifying patterns and relationships between multiple variables at once.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Creating scatter plot
sns.scatterplot(x=x, y=y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.


To plot a line plot using the "fmri" dataset from Seaborn's example datasets, specifically showing the relationship between the "timepoint" and "signal" for different "events" and "regions", follow these steps:

Import Libraries: Import the necessary libraries (seaborn and matplotlib.pyplot).
Load the Dataset: Use sns.load_dataset() to load the "fmri" dataset into a DataFrame.
Plotting: Utilize sns.lineplot() to create the line plot, setting "timepoint" as the x-axis (x) and "signal" as the y-axis (y), while differentiating the lines by "event" and using different colors/styles for different "regions".
Here's how you can implement this:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Plotting the line plot
plt.figure(figsize=(10, 6))  # Set the figure size (width, height)

# Use seaborn lineplot to create the plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Customize the plot appearance
plt.title('FMRI Signal Over Time by Event and Region')  # Set the plot title
plt.xlabel('Timepoint')  # Label for the x-axis
plt.ylabel('Signal')  # Label for the y-axis
plt.legend(title='Event')  # Add legend with title for events

# Display the plot
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

To plot two box plots using the "titanic" dataset from Seaborn's example datasets, specifically showing the distribution of "age" and "fare" across different passenger classes ("pclass"), you can follow these steps:

Import Libraries: Import the necessary libraries (seaborn and matplotlib.pyplot).
Load the Dataset: Use sns.load_dataset() to load the "titanic" dataset into a DataFrame.
Plotting: Utilize sns.boxplot() to create separate box plots for "age" and "fare" grouped by "pclass".

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Plotting two box plots side by side
plt.figure(figsize=(12, 6))  # Set the figure size (width, height)

# Box plot for 'age' grouped by 'pclass'
plt.subplot(1, 2, 1)  # Create subplot (1 row, 2 columns, 1st subplot)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

# Box plot for 'fare' grouped by 'pclass'
plt.subplot(1, 2, 2)  # Create subplot (1 row, 2 columns, 2nd subplot)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

# Adjust layout and display the plots
plt.tight_layout()  # Adjust subplot parameters to give specified padding
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

To plot a histogram for the 'price' column of the "diamonds" dataset from Seaborn, and use the hue parameter to differentiate the histograms by the 'cut' column, you can follow these steps:

Import Libraries: Import the necessary libraries (seaborn and matplotlib.pyplot).
Load the Dataset: Use sns.load_dataset() to load the "diamonds" dataset into a DataFrame.
Plotting: Utilize sns.histplot() to create the histogram, setting 'price' as the variable for the histogram and using hue='cut' to differentiate histograms by the 'cut' column.

Here's how you can implement this:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Plotting histogram with hue for 'cut'
plt.figure(figsize=(10, 6))  # Set the figure size (width, height)

# Create a histogram with hue for 'cut'
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack', bins=30)
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

To plot a pair plot (scatterplot matrix) for the "iris" dataset from Seaborn and use the hue parameter to differentiate the scatterplots by the "species" column, you can follow these steps:

Import Libraries: Import the necessary libraries (seaborn and matplotlib.pyplot).
Load the Dataset: Use sns.load_dataset() to load the "iris" dataset into a DataFrame.
Plotting: Utilize sns.pairplot() to create the pair plot, setting hue='species' to differentiate scatterplots by the 'species' column.
Here's how you can implement this:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris_data = sns.load_dataset("iris")

# Plotting pair plot with hue for 'species'
plt.figure(figsize=(8, 8))  # Set the figure size (width, height)

# Create a pair plot with hue for 'species'
sns.pairplot(data=iris_data, hue='species', markers=['o', 's', 'D'])

# Adjust plot title
plt.title('Pair Plot of Iris Dataset by Species')

# Display the plot
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to prepare for heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Plotting heatmap
plt.figure(figsize=(10, 8))  # Set the figure size (width, height)

# Create heatmap using seaborn
sns.heatmap(flights_pivot, cmap='Blues', annot=True, fmt='d', linewidths=1)

# Customize plot title and labels
plt.title('Number of Passengers by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

# Display the plot
plt.show()
```
