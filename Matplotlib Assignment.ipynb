Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular plotting library in Python used for creating static, interactive, and animated visualizations in a variety of formats. It provides a flexible framework for generating plots, charts, histograms, and other types of graphical representations from data.

Here are five common plots that can be created using the Pyplot module of Matplotlib:

Line Plot: This is a basic plot type used to visualize the relationship between two variables by connecting data points with straight lines.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]


plt.figure(figsize=(3,2))
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```python
# Scatter Plot: This plot displays individual data points as markers (dots)
# to show the relationship between two variables.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(3,2))
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
# Histogram: To show the distribution of data, use plt.hist(data, bins),
# which creates a histogram plot.

import matplotlib.pyplot as plt

x = [1, 2, 3,5,3,3,9, 4,2, 1,7,5]
# y = [2, 4, 6, 8, 10]

plt.figure(figsize=(3,2))
plt.hist(x)
plt.xlabel('X-Range')
plt.ylabel('Y-Frequency')
plt.title('Histogram Plot')
plt.show()
```

```python
# Bar Plot: For comparing categories of data, use plt.bar(x, height), which creates a bar chart.
import matplotlib.pyplot as plt

x = [1, 2, 2,4, 1, 7]
y = [2, 6,6,4, 8, 7]

plt.figure(figsize=(3,2))
plt.bar(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Plot')
plt.show()
```

```python
# Pie Chart: To represent proportions or percentages, use plt.pie(data, labels), which creates a pie chart.

import matplotlib.pyplot as plt

# Sample data for pie chart
labels = ['Apples', 'Oranges', 'Bananas', 'Mangoes']
sizes = [30, 25, 20, 25]  # Percentage distribution of each category

# Colors for each section of the pie chart
colors = ['gold', 'orange', 'lightgreen', 'lightcoral']

# Explode the 'Apples' section (pull it out)
explode = (0.1, 0, 0, 0)  # Explode the first slice (i.e., 'Apples')

# Plotting the pie chart
plt.figure(figsize=(4, 3))  # Set the figure size
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('Fruit Distribution')  # Set the title of the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the pie chart
plt.show()
```

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

Ans: A scatter plot is a type of data visualization that displays individual data points as markers (dots) on a two-dimensional plane.
Each point represents the values of two variables, one plotted along the x-axis and the other along the y-axis.
Scatter plots are useful for visualizing the relationship between two continuous variables, identifying patterns, clusters, or trends in the data, and assessing the correlation between variables.

Now, let's use the provided code to generate data for x and y and then create a scatter plot using Matplotlib with titles, labels, and markers.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for x and y
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plotting the scatter plot
plt.figure(figsize=(8, 6))  # Set the figure size

# Create a scatter plot
plt.scatter(x, y, color='#00CFFF', alpha=0.7, edgecolors='black', s=1000)

# Add title and labels
plt.title('Scatter Plot of Random Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure. This allows you to display different plots side by side or in a grid layout, making it easier to compare multiple visualizations at once.

The general syntax for subplot() is:
plt.subplot(num_rows, num_cols, plot_index)

num_rows: Number of rows in the subplot grid.
num_cols: Number of columns in the subplot grid.

plot_index: Index (starting from 1) of the subplot you want to create.

Now, let's draw four line plots using the subplot() function based on the provided data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.figure(figsize=(12, 10))  # Set the overall figure size

# Subplot 1
plt.subplot(2, 2, 1)  # (rows, columns, plot_index)
plt.plot(x1, y1, color='blue')
plt.title('Line Plot 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2, color='green')
plt.title('Line Plot 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3, color='red')
plt.title('Line Plot 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4, color='purple')
plt.title('Line Plot 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Adjust layout
plt.tight_layout()

# Display the subplots
plt.show()
```

Explanation of the code:

plt.figure(figsize=(12, 10)): Creates a new figure with a specified figure size (width=12 inches, height=10 inches) for better visualization of subplots.

plt.subplot(2, 2, 1), plt.subplot(2, 2, 2), plt.subplot(2, 2, 3), 

plt.subplot(2, 2, 4): Creates four subplots arranged in a 2x2 grid.

plt.plot(x, y, color): Plots the specified data (x, y) with the given color.

plt.title('Line Plot 1'), plt.title('Line Plot 2'), etc.: Sets titles for each subplot.

plt.xlabel('X-axis'), plt.ylabel('Y-axis'): Sets labels for the x-axis and y-axis of each subplot.
plt.tight_layout(): Automatically adjusts subplot parameters to fit into the figure area without overlapping.
plt.show(): Displays the figure containing all the subplots.

This code will generate four line plots (Line Plot 1, Line Plot 2, Line Plot 3, Line Plot 4) arranged in a 2x2 grid, each showing different data (x and y values). Adjust the colors, titles, labels, and other parameters as needed to customize the appearance of your subplots.

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

Solution: A bar plot is a type of data visualization that represents categorical data with rectangular bars. The length or height of each bar is proportional to the value it represents. Bar plots are commonly used to compare and visualize the quantities of different categories or groups.

Why is a bar plot used?

Comparison: Bar plots are effective for comparing the quantities or values of different categories.
Categorical Data: They are suitable for displaying categorical data where each bar represents a category or group.
Visual Clarity: Bar plots provide a clear visual representation of data, making it easy to interpret and identify trends or differences.
Now, let's create a bar plot and a horizontal bar plot using the provided data (company and profit).

Bar Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.bar(company, profit, color='skyblue')  # Create a bar plot
plt.title('Profit of Companies')  # Set the title
plt.xlabel('Company')  # Set the x-axis label
plt.ylabel('Profit (in billions)')  # Set the y-axis label
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Show grid lines on the y-axis

# Display the bar plot
plt.show()
```

```python
# Horizontal Bar Plot

import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.barh(company, profit, color='salmon')  # Create a horizontal bar plot
plt.title('Profit of Companies')  # Set the title
plt.xlabel('Profit (in billions)')  # Set the x-axis label
plt.ylabel('Company')  # Set the y-axis label
plt.grid(axis='x')  # Show grid lines on the x-axis

# Display the horizontal bar plot
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

Ans: A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It provides a concise way to visualize the spread and skewness of the data, including potential outliers.

Why is a box plot used?

Summary of Distribution: A box plot summarizes the key characteristics of a dataset, including its central tendency, spread, and skewness.
Identifying Outliers: It helps identify potential outliers in the dataset, which are points that fall significantly outside the range of typical values.
Comparison: Box plots are useful for comparing distributions of different datasets or different groups within a dataset.
Now, let's create a box plot using the provided data (box1 and box2).

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)  # Mean=100, Standard Deviation=10, Size=200
box2 = np.random.normal(90, 20, 200)   # Mean=90, Standard Deviation=20, Size=200

# Combine data into a list
data = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.boxplot(data, labels=['Box 1', 'Box 2'], patch_artist=True)  # Create a box plot
plt.title('Box Plot of Box 1 and Box 2')  # Set the title
plt.xlabel('Box')  # Set the x-axis label
plt.ylabel('Values')  # Set the y-axis label
plt.grid(True)  # Show grid lines

# Display the box plot
plt.show()
```
